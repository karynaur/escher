```
# load the image from disk
image = cv2.imread('matematicos_final.jpg')
# show image
cv2_imshow(image)
```

```
#https://www.pyimagesearch.com/2017/01/02/rotate-images-correctly-with-opencv-and-python/
#Rotate
angle = 157.62
#rotated = imutils.rotate(image, angle) #function that cuts corners
#cv2_imshow(rotated)
#ensure the entire image is bound inside the window and none is cut off.
rotated2 =imutils.rotate_bound(image, angle) #function that preserves the size
cv2_imshow(rotated2)
```

```
# Resize rotated image
scale_percent = 100-22.58
width = int(rotated2.shape[1] * scale_percent / 100)
height = int(rotated2.shape[0] * scale_percent / 100)
dim = (width, height)  
# resize image
resized = cv2.resize(rotated2, dim, interpolation = cv2.INTER_AREA)
cv2_imshow(resized)
```

```
#https://www.pyimagesearch.com/2017/01/02/rotate-images-correctly-with-opencv-and-python/
#detect edges
#image_filtered = cv2.Canny(image, 20, 100) #edges
image_filtered = cv2.Canny(image, 400, 550) #edges
cv2_imshow(image_filtered)
```
